```python
"""
Created on Wed Dec 11 01:12:42 2024

@author: Dr Ambi

Developed by Dr Ambi, copyright@Dr Ambi

#Tutorial to visualize relationship between a categorical variables and a continuous variable  

#Dataset: Auto MPG, Avialble from UCI Machine Learning Repository
#This car dataset is about fuel consumption in miles per gallon
#Source: https://archive.ics.uci.edu/dataset/9/auto+mpg

References: Python Pandas API: https://pandas.pydata.org/docs/reference/index.html

Seaborn catplot: Function to draw categorical plots
https://seaborn.pydata.org/generated/seaborn.catplot.html

Seaborn.stripplot
Function to draw categorical scatterplot using jitter to reduce overplotting
https://seaborn.pydata.org/generated/seaborn.stripplot.html#seaborn.stripplot

Seaborn heatmap function: This function plots tabular data as a color-encoded matrix.
https://seaborn.pydata.org/generated/seaborn.heatmap.html

Pandas cut function: This function is used to convert a continuous variable to a categorical variable
https://pandas.pydata.org/docs/reference/api/pandas.cut.html

Pandas crosstab function: Computes a frequency table of the factors 
https://pandas.pydata.org/docs/reference/api/pandas.crosstab.html

"""

#Dataset: Auto MPG, Avialble from UCI Machine Learning Repository
#This car dataset is about fuel consumption in miles per gallon
#Source: https://archive.ics.uci.edu/dataset/9/auto+mpg

import pandas as pd
import seaborn as sns

df=sns.load_dataset('mpg')
df.info()

#Below code means create a new variable mpgCategory, set its value to low, medium or high based on values of column mpg
#inf stands for infinity

df['mpgCategory']=pd.cut(df['mpg'], [0,15,30, float('inf')], labels =['low','medium','high'])

frequency=df['mpgCategory'].value_counts()

sns.catplot(data=df, x='mpgCategory', y='model_year', kind='strip', jitter=True)

sns.catplot(data=df, x='cylinders', y='model_year', kind='strip', jitter=True)
```
